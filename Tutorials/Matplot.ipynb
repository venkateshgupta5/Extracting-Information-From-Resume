# Cos

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.linspace(0,20,num=20)
y = np.cos(a)
plt.plot(a, y)
plt.show()
```

```python
plt.xlabel("Time")
plt.ylabel("Distance")
plt.title("My First Graph")
x = np.linspace(0,20,num=100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```
